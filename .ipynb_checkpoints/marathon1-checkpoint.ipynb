```python
# (A.) KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
n = marathon[['Age','M/F','Overall','Gender','Division','Country']]
```

```python
X = n[['Age', 'Overall','Gender','Division']]
y = n['Country']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=99)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
y_pred_class = knn.predict(X_test)
print((metrics.accuracy_score(y_test, y_pred_class)))
```

```python
knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(X_train, y_train)
y_pred_class = knn.predict(X_test)
print((metrics.accuracy_score(y_test, y_pred_class)))
```

```python
# Model With Five Neighbors.
```

```python
knn5 = KNeighborsClassifier(n_neighbors=5)

knn5.fit(X_train, y_train)
knn5.score(X_test, y_test)
```

```python
most_freq_class = y_train.value_counts().index[0]

print(y_train.value_counts())
most_freq_class
```

```python
y_test.value_counts()[most_freq_class] / len(y_test)
```

```python
knn.predict_proba(X)
```

```python
scores = []
for k in range(1,100):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X,y)
    pred = knn.predict(X)
    score = float(sum(pred == y)) / len(y)
    scores.append([k, score])
```

```python
data = pd.DataFrame(scores,columns=['k','score'])
data.plot.line(x='k',y='score');
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
predic = ['Age','M/F']
Xx = marathon[predic]
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xx, y, test_size=0.4,random_state=12345)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
lr = LinearRegression()
type(lr)
```

```python
model = lr.fit(X_train, y_train)
print(lr.intercept_)
```

```python
# coefficient
```

```python
print(lr.coef_)
```

```python
y_predicted = lr.predict(X_test)
y_predicted[:21]
```

```python
# MAE, MSE & RSME
```

```python
from sklearn import metrics

print('Mean-AbsoluteError:', metrics.mean_absolute_error(y_test, y_predicted))
print('Mean-Squared Error:', metrics.mean_squared_error(y_test, y_predicted))
print('Root-Mean-Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_predicted)))
```
